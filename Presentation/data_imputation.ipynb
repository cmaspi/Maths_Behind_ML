```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ppca import PPCA
from scipy import linalg

sns.set()
```

```python
sig = np.eye(3)
sig[(0, 1), (1, 0)] = 0.5
sig[(2, 1), (1, 2)] = 0.7

X_homo = np.random.multivariate_normal(np.zeros(3),
                                       sig,
                                       size = 100) 
```

```python
sig
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X_homo[:,0], X_homo[:,1], X_homo[:, 2], color='black')
plt.show()
```

```python
n_components = 2
ppca_model = PPCA(n_components=n_components)
ppca_model.fit(X_homo)
```

```python
arr = X_homo[1]
missing_indices = np.array((1,))
print(arr[missing_indices])

arr[missing_indices] = np.nan

arr_impute = ppca_model.impute_data(arr)
print(arr_impute[missing_indices])
```
